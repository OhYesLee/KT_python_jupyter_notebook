```python
import tensorflow as tf
from tensorflow import keras
```

```python
## Functional API
# 1번. 세션 클리어
keras.backend.clear_session()

# 2번. 레이어 사슬처럼 엮기
il = keras.layers.Input(shape=(5,5,3))
cl = keras.layers.Conv2D(       filters=1,                 # Conv2D를 통해 제작하려는 Feature map의 수
                                kernel_size=(3,3),         # filter size
                                strides=(1,1),             # filter가 얼마나 촘촘히 훑을 것인가 (sliding window)
                                # padding='same',          # filter가 훑기 전에 상하좌우로 픽셀을 덧붙임
                                activation='relu')(il)     # activation 주의!!!

pl = keras.layers.MaxPool2D(pool_size=(2,2), strides=(1,1))(cl)

# 3번. 모델의 시작/끝 지정
model = keras.models.Model(il, pl)

model.summary()
```
