### [Step2] 가설설정
#### '가설' 이란 단어를 생각해보면, 우리가 직접 가설을 만들어야 할것 같지만 통계에서 분석기법별 귀무가설/대립가설은 정해져있습니다.
- 귀무가설은 차이가 없다, 연관성이 없다, 효과가 없다.
- 대립가설은 차이가 있다, 연관성이 있다, 효과가 있다. 라고 간단하게 이해하시면 좋습니다.

#### 간단한 예로, 온도와 따릉이 대여량간 가설을 수립 해본다면

    - 귀무가설 : 온도와 따릉이 간에는 연관성이 없다.
    - 대립가설 : 온도와 따릉이 간에는 연관성이 있다.

로 가설이 자연스레 수립 됩니다.

5. 변수의 분포 확인

```python
#변수의 분포를 확인해보세요. 시각화
```

```python
def graph(var):
    plt.figure(figsize = (20,15))
    plt.subplot(3,1,1)
    sns.histplot(seoul_bike1[var], #data = seoul_bike1,
                      bins=20, ## bin 개수
                      kde=True, ## 밀도 함수 곡선 추가
                      element='step', ## 히스토그램 표시형식,
                      stat='percent'#, ## y값 계산 방식 count
                      #cumulative=False ## True인 경우 누적 분포 형태로 그림
                )

    plt.subplot(3,1,2)
    sns.kdeplot(seoul_bike1[var])

    plt.subplot(3,1,3)
    sns.boxplot(x= seoul_bike1[var],data = seoul_bike1)

    plt.tight_layout()
    plt.show()
```

```python
def graph2(var,target):
    plt.figure(figsize = (20,15))
    plt.subplot(3,1,1)
    sns.histplot(x= seoul_bike1[var] , y = seoul_bike1[target], 
                      bins=20, ## bin 개수
                      kde=True, ## 밀도 함수 곡선 추가
                      element='step', ## 히스토그램 표시형식,
                      stat='percent'#, ## y값 계산 방식 count
                      # cumulative=False ## True인 경우 누적 분포 형태로 그림
                )

    plt.subplot(3,1,2)
    sns.kdeplot(x= seoul_bike1[var] , y = seoul_bike1[target])

    plt.subplot(3,1,3)
    sns.boxplot(x= seoul_bike1[var] , y = seoul_bike1[target])

    plt.tight_layout()
    plt.show()
```

```python
graph2('시간','시간에 따른 따릉이 대여 수')
```

```python
graph('시간')
```

```python
target = '시간에 따른 따릉이 대여 수'
```

```python
def corrPlot(var, target):
#     sns.scatterplot(x=var1, y = target1, data = seoul)
    sns.regplot(x=var, y=target, data=seoul_bike1)  
    plt.title(f'{var} 와(과) {target} 의 상관관계')
    plt.show()
    
    result = spst.pearsonr(seoul_bike1[var], seoul_bike1[target])
    print(f'상관계수 : {result[0]}, p-value : {result[1]}')
```

```python
corrPlot('시간',target)
```

```python
plt.figure(figsize = (8, 8))
sns.heatmap(seoul_bike1.corr(), 
            annot = True,            # 숫자(상관계수) 표기 여부
            fmt = '.3f',             # 숫자 포멧 : 소수점 3자리까지 표기
            cmap = 'RdYlBu_r',       # 칼라맵
            vmin = -1, vmax = 1)     # 값의 최소, 최대값값
plt.show()
```
