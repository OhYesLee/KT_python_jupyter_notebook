## Quick Linear Regression!

```
## 라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(20))
y = x * 2 - 1 

print(x)
print(y)
```

```
x.shape, y.shape
```

```
## Sequential API
# 1. 세션 클리어 : 이전에 사용했던 모델이 있다면 없애줘!
keras.backend.clear_session()

# 2. 모델 발판 생성
model = keras.models.Sequential()

# 3. 모델 레이어 쌓기
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1) )

# 4. 컴파일
model.compile(loss='mse', optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
y_pred = model.predict(x)
```

```
y_pred.reshape(-1)
```

```
y
```

```
## 라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(20)) 
y = x * (-3) + 10

print(x)
print(y)
```

```
x.shape, y.shape
```

```
## 모델링 : Sequential API
# 세션 클리어 : 기존의 모델이 메모리에 남아있다면 지워줘!
keras.backend.clear_session()

# 모델을 쌓을 발판 생성
model = keras.models.Sequential()

# 모델 블록 조립
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1) )

# 컴파일
model.compile(loss='mse', optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=2)
```

```
y_pred = model.predict(x)
```

## Quick Logistic Regression!

```
## 라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras
```

```
x = np.array(range(20)) 
y = np.array([0]*10 + [1]*10)

print(x)
print(y)
```

```
x.shape, y.shape
```

```
## 모델링 : Sequential API
# 1. 세션 클리어 : 기존의 모델이 남아있다면 지워줘.
keras.backend.clear_session()

# 2. 레이어 발판 생성
model = keras.models.Sequential()

# 3. 레이어 블록 조립
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1, activation='sigmoid') )

# 4. 컴파일
model.compile(loss='binary_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```
# 학습
model.fit(x, y, epochs=10, verbose=1)
```

```
# 예측값
y_pred = model.predict(x)
```

```
y_pred[:5]
```

```
y[:5]
```

```
## 라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(40)) 
y = np.array([0]*20 + [1]*20)
print(x)
print(y)
```

```
x.shape, y.shape
```

```
## 모델링 : Sequential API
# 1번. 청소
keras.backend.clear_session()

# 2번. 모델 발판 생성
model = keras.models.Sequential()

# 3번. .add()
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1, activation='sigmoid') )

# 4번. .compile()
model.compile(loss='binary_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```
# 학습
model.fit(x, y, verbose=1, epochs=10)
```

```
y_pred = model.predict(x)
```
