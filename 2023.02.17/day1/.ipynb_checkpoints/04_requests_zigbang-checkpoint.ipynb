### Zigbang 원룸 매물 데이터 수집
- 동이름 -> 매물 데이터

```python
import pandas as pd
import requests
```

## 시크릿모드로 사용하면 좋음

#### Process
    - 동이름으로 위도 경도 구하기
    - 위도 경도로 geohash 알아내기
    - geohash로 매물 아이디 가져오기
    - 매물 아이디로 매물 정보 가져오기

#### 1. 동이름으로 위도 경도 구하기

```python
# 1. URl
address = '개포동' 
url = f'https://apis.zigbang.com/v2/search?leaseYn=N&q={address}&serviceType=원룸'

# 2. request > response : json(str)
response = requests.get(url)
response

# 3. json(str) > lat, lng
data = response.json()['items'][0]
lat, lng = data['lat'],data['lng']
lat,lng
```

#### 2. 위도 경도로 geohash 알아내기
- geohash2 : pip install geohash2
- 터미널에서 실행해야한다.

```python
import geohash2
# 변수명이 중복되면 안된다 절대로!!!
```

```python
# precision : 클수록 영역이 작아짐
geohash = geohash2.encode(lat,lng,precision=5)
geohash
```

#### 3. geohash로 매물 아이디 가져오기

```python
# \ + enter 하면 개행 다음으로 내려감
# decode안하고 우클릭 copy value해도 한글로 변환돼요!
# url = 'https://apis.zigbang.com/v2/items?deposit_gteq=0&domain=zigbang&geohash={geohash}\
# wydm5&needHasNoFiltered=true&rent_gteq=0&sales_type_in=전세|월세&service_type_eq=원룸'
# response_re = requests.get(url) 
# response_re
```

```python
data = response.json()['items']
ids = [item['item_id'] for item in data]
len(ids),ids[:5]
```

#### 4. 매물 아이디로 매물 정보 가져오기

```python
url = 'https://apis.zigbang.com/v2/items/list'
params = {
    'domain' : 'zigbang',
    'withCoalition' : 'true',
    'item_id' : ids[:900], # 아이템 데이터의 갯수를 999개까지 사용 가능
}
reponse = requests.post(url,params)
reponse
```
