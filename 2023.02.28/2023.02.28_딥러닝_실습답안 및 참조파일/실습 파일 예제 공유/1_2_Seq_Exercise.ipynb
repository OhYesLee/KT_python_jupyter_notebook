```
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
```

##Linear Regression on 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
x = boston.data
y = boston.target

x.shape, y.shape
```

```
## 모델링 : Sequential API
# 1번. 세션 클리어 : 이전 모델이 남아있다면 날려줘!
keras.backend.clear_session()

# 2번. 모델 발판 생성 : 순차적으로 쌓을 발판!
model = keras.models.Sequential()

# 3번. 모델 블록 조립 : .add()
model.add( keras.layers.Input(shape=(13,)) )
model.add( keras.layers.Dense(1, activation='linear') )

# 4번. 컴파일
model.compile(loss='mse', optimizer='adam')
```

```
# 학습
model.fit(x, y, epochs=10, verbose=1)
```

```
# 예측값 생성
y_pred = model.predict(x)
```

```
y_pred[:10].reshape(-1)
```

```
y[:10]
```

```
boston = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/boston.csv')
```

```
boston.head()
```

```
x = boston.iloc[ : , :-1]
y = boston.iloc[ : , -1 ]
```

```
x.shape, y.shape
```

```
## 모델링 2 : Sequential API
# 세션 클리어
keras.backend.clear_session()

# 시퀀셜하게 쌓을거야!
model = keras.models.Sequential()

# 레이어 조립
model.add( keras.layers.Input(shape=(12,)) )
model.add( keras.layers.Dense(1, activation='linear') )

# 컴파일
model.compile(loss=keras.losses.mean_squared_error, optimizer=keras.optimizers.Adam() )
```

```
# 학습
model.fit(x, y, epochs=10, verbose=1)
```

```
# 예측
y_pred = model.predict(x)
y_pred[:5].reshape(-1)
```

```
y[:5]
```

## Logistic Regression 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_breast_cancer
```

```
cancer = load_breast_cancer()
```

```
x = cancer.data
y = cancer.target

# 모양 확인해볼 것
```

```
x.shape, y.shape
```

```
## 모델링 : Sequential API
# 1. 세션 클리어
keras.backend.clear_session()

# 2. 모델 발판 생성
model = keras.models.Sequential()

# 3. 모델 블록 조립
model.add(keras.layers.Input(shape=(30,)) )
model.add(keras.layers.Dense(1, activation='sigmoid') )

# 4. 컴파일
model.compile(loss='binary_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```
## 학습
model.fit(x, y, epochs=10, verbose=1)
```

```
## 예측값 생성
y_pred = model.predict(x)
```
