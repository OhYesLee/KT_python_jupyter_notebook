### 1) **본인의 CustomData yaml 다운로드**

- CustomData yaml 사전 작업 필요

```python
!wget -O /content/yolov5/data/cat_dog_person.yaml https://raw.githubusercontent.com/DrKAI/CV/main/cat_dog_person.yaml
```

### 2) Pretrained weights 다운로드

![Pretrained weights](https://github.com/DrKAI/CV/raw/main/UltraLytics_manual/yolov5_pretrained.png)

[Pretrained weights Page](https://github.com/ultralytics/yolov5)

```python
!mkdir /content/yolov5/pretrained
!wget -O /content/yolov5/pretrained/yolov5s.pt https://github.com/ultralytics/yolov5/releases/download/v6.2/yolov5s.pt
```

### 3) train.py 실행

- 명령어 도움말 : python train.py -h

```python
!cd yolov5; python train.py \
    --data '/content/yolov5/data/cat_dog_person.yaml' \
    --cfg '/content/yolov5/models/yolov5s.yaml' \
    --weights '/content/yolov5/pretrained/yolov5s.pt' \
    --epochs 100 \
    --patience 7 \
    --img 640 \
    --project 'trained' \
    --name 'train_cat_dog_person' \
    --exist-ok
    # --device cpu
```

### 4) **테스트용 동영상 다운로드**

```python
!mkdir /content/yolov5/data/videos
```

```python
!wget -O /content/yolov5/data/videos/cat_babysitter.mp4 https://github.com/DrKAI/CV/raw/main/example/Cat_Babysitter.mp4
!wget -O /content/yolov5/data/videos/cat_dog1.mp4 https://github.com/DrKAI/CV/raw/main/example/Cat_Dog1.mp4
!wget -O /content/yolov5/data/videos/cat_dog2.mp4 https://github.com/DrKAI/CV/raw/main/example/Cat_Dog2.mp4
!wget -O /content/yolov5/data/videos/cat_dog_comfort.mp4 https://github.com/DrKAI/CV/raw/main/example/Cat_Dog_comfort.mp4
!wget -O /content/yolov5/data/videos/cat_man.mp4 https://github.com/DrKAI/CV/raw/main/example/Cat_Man.mp4
!wget -O /content/yolov5/data/videos/cat_puppy.mp4 https://github.com/DrKAI/CV/raw/main/example/Cat_Puppy.mp4
!wget -O /content/yolov5/data/videos/kitten_dog.mp4 https://github.com/DrKAI/CV/raw/main/example/Kitten_Dog.mp4
!wget -O /content/yolov5/data/videos/cat_dog_hiking.mp4 https://github.com/DrKAI/CV/raw/main/example/Cat_Dog_Hiking.mp4
```

### 5) detect.py 실행

- 명령어 도움말 : python detect.py -h

```python
!cd yolov5; python detect.py \
    --weights '/content/yolov5/trained/train_cat_dog_person/weights/best.pt' \
    --source '/content/yolov5/data/videos' \
    --project '/content/yolov5/detected' \
    --name 'video_cat_dog_person' \
    --img 640 \
    --conf-thres 0.25 \
    --iou-thres 0.5 \
    --line-thickness 2 \
    --exist-ok 
    # --device CPU
```
