### 1) **본인의 CustomData yaml 다운로드**

- CustomData yaml 사전 작업 필요

```python
!wget -O /content/yolov5/data/cat_dog_person.yaml https://raw.githubusercontent.com/DrKAI/CV/main/cat_dog_person.yaml
```

### 2) Pretrained weights 다운로드

![Pretrained weights](https://github.com/DrKAI/CV/raw/main/UltraLytics_manual/yolov5_pretrained.png)

[Pretrained weights Page](https://github.com/ultralytics/yolov5)

```python
!mkdir /content/yolov5/pretrained
!wget -O /content/yolov5/pretrained/yolov5m.pt https://github.com/ultralytics/yolov5/releases/download/v6.2/yolov5m.pt
```

### 3) train.py 실행

- 명령어 도움말 : python train.py -h

```python
!cd yolov5; python train.py \
    --data '/content/yolov5/data/cat_dog_person.yaml' \
    --cfg '/content/yolov5/models/yolov5m.yaml' \
    --weights '/content/yolov5/pretrained/yolov5m.pt' \
    --epochs 100 \
    --patience 10 \
    --img 640 \
    --project 'trained' \
    --name 'train_CatDogPerson' \
    --exist-ok
    # --device cpu # 학습은 GPU로 하는 것이 빠르다.
```

### 4) **테스트용 이미지 다운로드**

```python
!wget -O /content/yolov5/data/images/kitten_person1.jpg https://github.com/DrKAI/CV/raw/main/example/kitten_person1.jpg
!wget -O /content/yolov5/data/images/kitten_person2.webp https://github.com/DrKAI/CV/raw/main/example/kitten_person2.webp
!wget -O /content/yolov5/data/images/kitten_person3.webp https://github.com/DrKAI/CV/raw/main/example/kitten_person3.webp
!wget -O /content/yolov5/data/images/kitten_person4.jpg https://github.com/DrKAI/CV/raw/main/example/kitten_person4.jpg

!wget -O /content/yolov5/data/images/kitten_puppy2.webp https://github.com/DrKAI/CV/raw/main/example/kitten_puppy2.webp
!wget -O /content/yolov5/data/images/kitten_puppy3.webp https://github.com/DrKAI/CV/raw/main/example/kitten_puppy3.webp
!wget -O /content/yolov5/data/images/kitten_puppy4.jpg https://github.com/DrKAI/CV/raw/main/example/kitten_puppy4.jpg
!wget -O /content/yolov5/data/images/kitten_puppy5.webp https://github.com/DrKAI/CV/raw/main/example/kitten_puppy5.webp
!wget -O /content/yolov5/data/images/kitten_puppy6.jpg https://github.com/DrKAI/CV/raw/main/example/kitten_puppy6.jpg

!wget -O /content/yolov5/data/images/puppy_person1.webp https://github.com/DrKAI/CV/raw/main/example/puppy_person1.webp
!wget -O /content/yolov5/data/images/puppy_person2.jpg https://github.com/DrKAI/CV/raw/main/example/puppy_person2.jpg
!wget -O /content/yolov5/data/images/puppy_person3.webp https://github.com/DrKAI/CV/raw/main/example/puppy_person3.webp
!wget -O /content/yolov5/data/images/puppy_person4.jpeg https://github.com/DrKAI/CV/raw/main/example/puppy_person4.jpeg
```

### 5) detect.py 실행

- 명령어 도움말 : python detect.py -h

```python
!cd yolov5; python detect.py \
    --weights '/content/yolov5/trained/train_CatDogPerson/weights/best.pt' \
    --source '/content/yolov5/data/images' \
    --project '/content/yolov5/detected' \
    --name 'img_CatDogPerson' \
    --img 640 \
    --conf-thres 0.6 \
    --iou-thres 0.5 \
    --line-thickness 2 \
    --exist-ok 
    # --device CPU
```
