```python
df = pd.DataFrame({'foo': ['one','one','one','two','two','two'],
                       'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                       'baz': [1, 2, 3, 4, 5, 6]})
df
```

```python
df2 = df.pivot(index='foo', columns='bar', values='baz')
```

```python
df.pivot(columns='bar')['baz']
```

```python
df3 = df2.reset_index()
df3
```

```python
df3.melt(id_vars=['foo'], value_vars=['A', 'B', 'C']).sort_values(by=['foo', 'bar'])
```
