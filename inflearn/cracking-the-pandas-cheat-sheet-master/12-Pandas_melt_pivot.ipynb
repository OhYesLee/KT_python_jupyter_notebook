```python
import pandas as pd
```

```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
df
```

```python
pd.melt(df, value_vars=['A', 'B', 'C']).rename(columns={
'variable' : 'var',
'value' : 'val'})
```

```python
df2 = pd.DataFrame({'foo': ['one','one','one','two','two','two'],
                    'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'baz': [1, 2, 3, 4, 5, 6]})
df2
```

```python
df3 = df2.pivot(index='foo', columns='bar', values='baz').reset_index()
df3
```

```python
df3.melt(
    id_vars=['foo'], 
    value_vars=['A', 'B', 'C']).sort_values(
    ['foo', 'bar']).rename(columns = {'value':'baz'})
```
